```python
sports = {
    "Atletismo": 1,
    "Baloncesto": 5,
    "Boxeo": 1,
    "Esgrima": 1,
    "Volleyball": 5,
    "Natación": 1,
    "Karate": 1,
    "Hockey (Hielo)": 6,
}

sport_points = {
    "Atletismo": 100,
    "Baloncesto": 20,
    "Boxeo": 100,
    "Esgrima": 100,
    "Volleyball": 20,
    "Natación": 100,
    "Karate": 100,
    "Hockey (Hielo)": 20,   
}

participants = {participantID: dict() for participantID in range(1, 135)}
points = {participantID: 0 for participantID in range(1, 135)}
```

```python
import random

teams = 15

for sport, teamsize in sports.items():
    # Get teams for each sport (competition)
    competitors = random.sample(sorted(participants), teamsize * teams)
    
    # Choose a random distribution of placements
    placements = random.sample(range(1,teams + 1), teams)
    
    # For each team assign a placement
    counter = -1
    for idx, competitor in enumerate(competitors):
        if idx % teamsize == 0:
            counter += 1
        
        participants[competitor][sport] = placements[counter]
```

```python
points = {participantID: 0 for participantID in range(1, 135)}
sport_points = {
    "Atletismo": 100,
    "Baloncesto": 20,
    "Boxeo": 100,
    "Esgrima": 100,
    "Volleyball": 20,
    "Natación": 100,
    "Karate": 100,
    "Hockey (Hielo)": 20,   
}

def getPercentage(placement):
    if placement == 1:
        return 1
    elif placement == 2:
        return 0.7
    elif placement == 3:
        return 0.6
    elif placement == 4:
        return 0.5
    elif placement == 5:
        return 0.3
    elif placement == 6:
        return 0.1
    else:
        return 0


for p_key in participants:
    for sport, placement in participants[p_key].items():
        points[p_key] += getPercentage(placement) * sport_points[sport]
```
